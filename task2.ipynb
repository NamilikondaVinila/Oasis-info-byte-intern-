```python
#datset link"https://www.kaggle.com/datasets/dgomonov/new-york-city-airbnb-open-data

import pandas as pd
import numpy as np

# Step 1: Data Loading
dataset = pd.read_csv(r"C:\Desktop\oasisintern\project3.csv")

# Step 2: Data Exploration
print(dataset.head())
print(dataset.info())

# Step 3: Missing Data Handling
# Check for missing values
missing_values = dataset.isnull().sum()
print("Missing Values:\n", missing_values)

# Handle missing data by imputing or removing
# Example: Impute missing values with mean for numeric columns
numeric_cols = dataset.select_dtypes(include=np.number).columns
dataset[numeric_cols] = dataset[numeric_cols].fillna(dataset[numeric_cols].mean())

# Step 4: Duplicate Removal
# Check for duplicates
duplicate_records = dataset.duplicated().sum()
print("Number of Duplicate Records:", duplicate_records)

# Remove duplicates
dataset.drop_duplicates(inplace=True)

# Step 5: Standardization (if required)
# Example: Convert string columns to lowercase
# Example: Convert date columns to datetime format

# Step 6: Outlier Detection (if required)
# Example: Detect and handle outliers using Z-score or IQR methods

# Step 7: Save Cleaned Dataset
dataset.to_csv("cleaned_dataset.csv", index=False)
```
